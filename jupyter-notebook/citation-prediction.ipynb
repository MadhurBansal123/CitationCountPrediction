```python
import networkx as nx
import requests
import json
```

```python
links = requests.get('http://localhost:8080/graphLink/getAllLinks')
print(links)
```

```python
linksObj = links.json()
```

```python
linksObj
```

```python
type(linksObj)
```

```python
linksObj[0]["toNodeId"]
```

```python
G = nx.DiGraph()

for i in linksObj:
    G.add_node(i["fromNodeId"])
```

```python
for i in linksObj:
    for j in i["toNodeId"]:
        G.add_edge(i["fromNodeId"],j)
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```

```python
from pyvis.network import Network
nt = Network('5000px', '5000px')
nt.from_nx(G)
nt.show('nx.png')
```
